```python
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nome = "victor"
url = f"https://servicodados.ibge.gov.br/api/v2/censos/nomes/{nome}"
params = {
    #  'sexo' : 'F'
    #, 'localidade' : 33
    }
response = requests.request(method="GET", url=url, params=params)
```

```python
response.json()
```

```python
response.url
```

```python
data_json = response.json()[0]['res']
```

```python
data_json
```

```python
nome_df = pd.DataFrame(data_json)
```

```python
ax = sns.barplot(data = nome_df, x = 'periodo', y = 'frequencia')
plt.xticks(rotation = 60)
ax.set(ylabel = f'frequencia nome "{nome}"')
```
