```python
import pandas as pd
import requests
```

## Requisições - Entendimento da API

```python
url = 'http://servicodados.ibge.gov.br/api/v3/noticias/'

params = {  'qtd' : 500
          , 'page': 1
          }

response = requests.request('GET', url = url, params = params)
```

```python
#@title Dados em formato tabular DataFrame
df = pd.DataFrame(response.json()['items'])
df.head()
```

## Paginação e Obtenção dos Dados

```python
url = 'http://servicodados.ibge.gov.br/api/v3/noticias/'

pagina = 1
params = {  'qtd' : 500
          , 'page': pagina}
noticiasDF = pd.DataFrame()

while pagina != 0:
    params['page'] = pagina

    resposta_json = requests.request('GET', url = url, params = params).json()

    num_item_inicial = resposta_json['showingFrom']
    num_item_final = resposta_json['showingTo']
    qtd_itens_exibidos = (num_item_final - num_item_inicial + 1)

    mensagem = f'''Analisando requisição da página {pagina}:
    Número item inicial: {num_item_inicial}
    Número item final: {num_item_final}
    Quantidade de itens exibidos: {qtd_itens_exibidos}\n
    '''

    print(mensagem)

    items = resposta_json['items']

    tempDF = pd.DataFrame(items)

    noticiasDF = pd.concat([noticiasDF, tempDF])

    pagina = resposta_json['nextPage']

```
